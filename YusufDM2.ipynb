Berikut adalah penjelasan untuk kode di atas:

Metode `isnull()` menghasilkan DataFrame yang berisi nilai boolean True atau False, di mana True menunjukkan bahwa nilai tersebut kosong (NaN), dan False menunjukkan bahwa nilai tersebut tidak kosong. Kemudian, method `sum()` digunakan untuk menghitung jumlah nilai True (yang berarti nilai kosong) di setiap kolom.

2. **Menampilkan jumlah nilai kosong**: Hasil dari kode tersebut adalah jumlah nilai kosong pada tiap kolom dalam dataset. Jika ada nilai kosong, kita akan mendapatkan jumlahnya; jika tidak ada nilai kosong, hasilnya akan menunjukkan bahwa setiap kolom memiliki jumlah nilai kosong yang nol.

Dengan melakukan pemeriksaan ini, kita dapat memastikan apakah dataset kita lengkap atau terdapat nilai yang kosong. Jika ada nilai kosong, langkah selanjutnya adalah menentukan strategi untuk menangani nilai kosong tersebut, seperti penghapusan baris atau kolom yang mengandung nilai kosong, atau pengisian nilai kosong dengan nilai yang sesuai.

####Outliers Values

```python
def detect_outliers(df):
    outliers = {}
    for col in df.columns:
        if df[col].dtype != 'object':
            Q1 = df[col].quantile(0.25)
            Q3 = df[col].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            outliers[col] = df[(df[col] < lower_bound) | (df[col] > upper_bound)].shape[0]
    return outliers

outliers = detect_outliers(df)
print("Jumlah outliers pada tiap kolom:")
print(outliers)
```

1. Mengecek Outliers: Dalam fungsi detect_outliers, kita melakukan iterasi pada setiap kolom dalam DataFrame. Untuk setiap kolom numerik, kita menghitung quartile pertama (Q1) dan quartile ketiga (Q3), kemudian menghitung jangkauan interquartile (IQR) sebagai selisih antara Q3 dan Q1. Kemudian, kita menentukan batas bawah dan batas atas untuk outliers dengan menggunakan faktor 1.5 kali IQR dari Q1 dan Q3. Data di luar jangkauan ini dianggap sebagai outliers.

2. Menampilkan Jumlah Outliers: Setelah menghitung jumlah outliers untuk setiap kolom, kita menampilkan hasilnya. Jika ada outliers, kita akan mendapatkan jumlahnya; jika tidak ada, hasilnya akan menunjukkan bahwa setiap kolom tidak memiliki outliers.

Dengan melakukan pemeriksaan ini, kita dapat mengetahui apakah ada outliers dalam dataset kita, dan ini dapat memberi kita wawasan tambahan tentang distribusi data kita.






####Duplicated Values

```python
def check_duplicates(df):
    duplicates = {}
    for col in df.columns:
        duplicates[col] = df.duplicated(subset=[col]).sum()
    return duplicates

duplicate_counts = check_duplicates(df)
print("Jumlah nilai duplikat pada tiap kolom:")
print(duplicate_counts)
```

Penjelasan:

1. **Mengecek Duplikat**: Dalam fungsi `check_duplicates`, kita melakukan iterasi pada setiap kolom dalam DataFrame. Untuk setiap kolom, kita menggunakan method `duplicated()` untuk memeriksa keberadaan nilai duplikat. Argument `subset=[col]` digunakan untuk membatasi pengecekan pada kolom tertentu. Method `sum()` kemudian digunakan untuk menghitung jumlah nilai duplikat dalam kolom tersebut.

2. **Menampilkan Jumlah Nilai Duplikat**: Setelah menghitung jumlah nilai duplikat untuk setiap kolom, kita menampilkan hasilnya. Jika ada nilai duplikat, kita akan mendapatkan jumlahnya; jika tidak, hasilnya akan menunjukkan bahwa setiap kolom tidak memiliki nilai duplikat.

Dengan melakukan pemeriksaan ini, kita dapat memastikan apakah ada nilai duplikat dalam dataset kita, dan ini dapat membantu kita dalam membersihkan dan mengolah data dengan lebih baik.Jangan lupa juga untuk berikan penjelasan di sini (baik ada nilai duplikat maupun gak ada nilai duplikat)
